```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

x = np.linspace(-5, +5, 101)
y1 = stats.t.pdf(x, 1)
y3 = stats.t.pdf(x, 3)
y10 = stats.t.pdf(x, 10)
y30 = stats.t.pdf(x, 30)
y100 = stats.t.pdf(x, 100) 
y = stats.t.pdf(x, 100000) 
y_norm = stats.norm.pdf(x) # 正規分布

plt.plot(x, y1, label="Freedom1") 
plt.plot(x, y3, label="Freedom 3") 
plt.plot(x, y10,  label="Freedom 10") 
plt.plot(x, y30,  label="Freedom 30") 
plt.plot(x, y100,  label="Freedom 100") 
plt.plot(x, y_norm,  label="Norm") 
plt.legend()
plt.show()
```
